HANDLING MISSING DATA

```python
ds['Age'].fillna(ds['Age'].median(),inplace=True)
ds['Embarked'].fillna(ds['Embarked'].mode()[0],inplace=True)
```

```python
ds.info()
```

AS CABIN COLUMN HAS MORE NULL VALUES,IT IS BETTER TO REMOVE THAT COLUMN FROM THE DATASET

```python
ds=ds.drop(['Cabin'],axis='columns')
```

```python
ds
```

```python
df=ds[['Survived','Pclass','Sex','Age']]
df
```

```python
df=pd.get_dummies(df,columns=['Sex'])
df
```

```python
df['Sex_female']=df['Sex_female'].astype(int)
df['Sex_male']=df['Sex_male'].astype(int)
```

```python
df
```

```python
x=df.drop(['Survived','Sex_male'],axis='columns')
x
```

```python
x=x.rename(columns={'Sex_female':'Gender'})
x   #(female=1 , male=0)
```

```python
y=df.Survived
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
for i in range(len(x_test)):
    if model.predict([x_test.iloc[i]])==1 :
        print("Survived")
    else :
        print("Not Survived")
```
